```python
# 12) Elabore um código que receba os dados de entrada a altura de uma pessoa, e informe qual seria o peso ideal, usando a seguinte fórmula: (72.7*altura) - 58

altura = float(input("Digite a altura da pessoa em metros (ex: 1.75): "))

peso_ideal = (72.7 * altura) - 58

print(f"Para uma pessoa com {altura:.2f} metros de altura, o peso ideal seria de {peso_ideal:.2f} kg.")
```

```python
# 11) Elabore um código que peça 2 números inteiros e um número real. Calcule e mostre:
# * o produto do dobro do primeiro com metade do segundo .
# * a soma do triplo do primeiro com o terceiro.
# * o terceiro elevado ao cubo.

numero_inteiro1 = int(input("Digite o primeiro número inteiro: "))
numero_inteiro2 = int(input("Digite o segundo número inteiro: "))
numero_real = float(input("Digite um número real: ").replace(',', '.'))

# a) o produto do dobro do primeiro com metade do segundo
resultado_a = (numero_inteiro1 * 2) * (numero_inteiro2 / 2)

# b) a soma do triplo do primeiro com o terceiro
resultado_b = (numero_inteiro1 * 3) + numero_real

# c) o terceiro elevado ao cubo
resultado_c = numero_real ** 3

print(f"a) O produto do dobro do primeiro com metade do segundo é: {resultado_a:.2f}")
print(f"b) A soma do triplo do primeiro com o terceiro é: {resultado_b:.2f}")
print(f"c) O terceiro elevado ao cubo é: {resultado_c:.2f}")
```

```python
# 10) Elabore um código que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.

celsius = float(input("Digite a temperatura em graus Celsius: "))

fahrenheit = (celsius * 9/5) + 32

print(f"A temperatura de {celsius}°C equivale a {fahrenheit:.2f}°F.")
```

```python
# 9) Elabore um código que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
# Onde C = 5 * ((F-32) / 9).

fahrenheit = float(input("Digite a temperatura em graus Fahrenheit: "))

celsius = 5 * ((fahrenheit - 32) / 9)

print(f"A temperatura de {fahrenheit}°F equivale a {celsius:.2f}°C.")
```

```python
# 8) Elabore um código que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês em uma mensagem amigável.

valor_por_hora = float(input("Quanto você ganha por hora? R$ "))
horas_trabalhadas_mes = float(input("Quantas horas você trabalhou no mês? "))

salario_total = valor_por_hora * horas_trabalhadas_mes

print(f"Olá! Seu salário total neste mês, considerando R$ {valor_por_hora:.2f} por hora e {horas_trabalhadas_mes} horas trabalhadas, é de R$ {salario_total:.2f}.")
```

```python
# 7) Elabore um código que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.

lado = float(input("Digite o comprimento do lado do quadrado: "))

area = lado ** 2
dobro_da_area = area * 2

print(f"A área do quadrado com lado {lado} é {area}.")
print(f"O dobro desta área é: {dobro_da_area}")
```

```python
# 6) Elabore um código que peça o raio de um círculo, calcule e mostre sua área.

import math

raio = float(input("Digite o raio do círculo: "))

area = math.pi * (raio ** 2)

print(f"A área do círculo com raio {raio} é: {area:.2f}")
```

```python
# 5) Elabore um código que converta metros para centímetros.

metros = float(input("Digite o valor em metros: "))

centimetros = metros * 100

print(f"{metros} metros equivalem a {centimetros} centímetros.")
```

```python
# 4) Elabore um código que peça as 4 notas bimestrais e mostre a média.

nota1 = float(input("Digite a primeira nota bimestral: "))
nota2 = float(input("Digite a segunda nota bimestral: "))
nota3 = float(input("Digite a terceira nota bimestral: "))
nota4 = float(input("Digite a quarta nota bimestral: "))

media = (nota1 + nota2 + nota3 + nota4) / 4

print(f"A média das notas bimestrais é: {media}")
```

```python
# 3) Elabore um código que peça dois números e imprima a soma.

numero1 = float(input("Por favor, digite o primeiro número: "))
numero2 = float(input("Por favor, digite o segundo número: "))

soma = numero1 + numero2

print(f"A soma dos números é: {soma}")
```
