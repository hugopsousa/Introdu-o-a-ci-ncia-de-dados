```python
def calcula_desconto(valor, porcentagem_desconto):
    desconto = valor * (porcentagem_desconto / 100)
    valor_final = valor - desconto
    return valor_final
```

```python
valor_produto = 100.00
percentual_desconto = 15
valor_com_desconto = calcula_desconto(valor_produto, percentual_desconto)

print(f"Valor original do produto: R${valor_produto:.2f}")
print(f"Desconto aplicado: {percentual_desconto}%")
print(f"Valor final com desconto: R${valor_com_desconto:.2f}")
```

```python
lista_de_numeros = [10, -5, 0, 7, -2, 0, 15, -8, 3, 0]

positivos = 0
negativos = 0
zeros = 0

for numero in lista_de_numeros:
    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Na lista {lista_de_numeros}:")
print(f"- Números Positivos: {positivos}")
print(f"- Números Negativos: {negativos}")
print(f"- Zeros: {zeros}")
```

```python
def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif 5 <= nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"
```

```python
nota_exemplo = 6.5
resultado = classifica_nota(nota_exemplo)
print(f"Com a nota {nota_exemplo}, o aluno está: {resultado}")

nota_exemplo_2 = 8.0
resultado_2 = classifica_nota(nota_exemplo_2)
print(f"Com a nota {nota_exemplo_2}, o aluno está: {resultado_2}")

nota_exemplo_3 = 4.9
resultado_3 = classifica_nota(nota_exemplo_3)
print(f"Com a nota {nota_exemplo_3}, o aluno está: {resultado_3}")
```
